v. explain the concept of type casting in Python with examples

ans. Type casting in Python is the process of converting one data type to another. It is useful when you need to perform operations that require matching types or to handle input/output conversions. Python provides built-in functions to achieve type casting.

 Types of Type Casting:

1. Implicit Type Casting**: This is automatically done by Python when performing operations that mix data types. Python converts the data type without user intervention.

   Example:
   x = 5       # int
   y = 2.5     # float
   result = x + y  # Python converts x to float
   print(result)   # Output: 7.5 (float)
   ```

2. Explicit Type Casting**: This is done manually by the programmer using Python's built-in functions like `int()`, `float()`, `str()`, `list()`, etc.

   Examples:
   - Integer to String Conversion:
     num = 123
     num_str = str(num)  # Convert integer to string
     print(num_str)      # Output: '123'
     print(type(num_str))  # Output: <class 'str'>
     ```

   - String to Integer Conversion:
     num_str = "456"
     num_int = int(num_str)  # Convert string to integer
     print(num_int)         # Output: 456
     print(type(num_int))   # Output: <class 'int'>
     ```

   - Float to Integer Conversion:
     num_float = 7.89
     num_int = int(num_float)  # Convert float to integer
     print(num_int)           # Output: 7 (decimal part is removed)
     print(type(num_int))     # Output: <class 'int'>
     ```

   - List to Tuple Conversion:
     my_list = [1, 2, 3]
     my_tuple = tuple(my_list)  # Convert list to tuple
     print(my_tuple)           # Output: (1, 2, 3)
     print(type(my_tuple))     # Output: <class 'tuple'>
     ```

 Importance of Type Casting:
- Compatibility: Ensures that operations between different data types are possible.
- Precision: Helps in preserving data types suitable for specific operations.
- Input Handling: Converts user input into the required data type for processing. 

Type casting is a crucial aspect of programming in Python, allowing for flexibility and control over how data is handled and manipulated.

vi.How do conditional statements work in Python? Illustrate with examples

Conditional statements in Python allow you to execute certain blocks of code based on whether a condition (or set of conditions) is true or false. They help control the flow of a program by making decisions.

 Types of Conditional Statements:

1. `if` statement: Executes a block of code if a specified condition is true.
2. `if-else` statement: Executes one block of code if the condition is true, otherwise executes another block.
3. `if-elif-else` statement: Allows checking multiple conditions, executing different blocks of code based on which condition is true.

### Syntax and Examples:

1. **`if` Statement**:

   age = 18
   if age >= 18:
       print("You are an adult.")
   ```
   - **Explanation**: Here, the code checks if `age` is greater than or equal to 18. If true, it prints "You are an adult."

2. `if-else` Statement:

   age = 16
   if age >= 18:
       print("You are an adult.")
   else:
       print("You are not an adult.")
   ```
   - **Explanation**: If the condition `age >= 18` is false (as `age` is 16), the program executes the code in the `else` block, printing "You are not an adult."

3. `if-elif-else` Statement:

   marks = 85
   if marks >= 90:
       print("Grade: A")
   elif marks >= 75:
       print("Grade: B")
   elif marks >= 60:
       print("Grade: C")
   else:
       print("Grade: D")
   ```
   - Explanation**: The code checks multiple conditions in sequence:
     - If `marks` is 90 or above, it prints "Grade: A".
     - If not, it checks if `marks` is 75 or above, then prints "Grade: B".
     - If not, it checks if `marks` is 60 or above, then prints "Grade: C".
     - If none of these conditions are true, it defaults to printing "Grade: D".

 How Conditional Statements Work:

1. **Evaluation**: The conditions in the `if`, `elif`, and `else` statements are evaluated as either `True` or `False`.
2. **Execution**: The block of code corresponding to the first true condition is executed. Once a true condition is found, the remaining conditions (if any) are not evaluated.
3. **Indentation**: Python uses indentation (spaces or tabs) to define the scope of the conditional blocks.

 Nested Conditional Statements:

Conditional statements can also be nested, meaning you can place an `if` statement inside another `if` statement to check additional conditions.


num = 10
if num > 0:
    print("Positive number")
    if num % 2 == 0:
        print("Even number")
    else:
        print("Odd number")
else:
    print("Non-positive number")
```

- Explanation: Here, the outer `if` checks if `num` is positive. Inside this block, another `if-else` checks if `num` is even or odd.

 Conclusion:
Conditional statements are fundamental in programming for decision-making. They enable a program to react differently based on the inputs or conditions encountered during execution.

vii. Descsibe the different types of loops in Python and their use cases with examples.

ans. Python provides two primary types of loops: `for` loops and `while` loops. Both are used to repeat a block of code multiple times, but they differ in how they control the iteration process.

 1. `for` Loop

The `for` loop is used to iterate over a sequence (such as a list, tuple, string, or range) and execute a block of code for each element in the sequence. It's typically used when the number of iterations is known or when iterating over a collection.

Syntax:


for variable in sequence:
    # code block
    
Example 1: Iterating over a List


fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
```

Output:

```
apple
banana
cherry
```

Example 2: Using `range()` Function

for i in range(5):
    print(i)
```

Output:

```
0
1
2
3
4
```

Use Cases:

- Iterating over elements in a list or tuple.
- Repeating a block of code a specific number of times.
- Iterating over characters in a string.

### 2. **`while` Loop**

The `while` loop executes a block of code as long as a specified condition is `True`. It is generally used when the number of iterations is not known beforehand and depends on dynamic conditions.

Syntax:


while condition:
    # code block
```

Example:


count = 0
while count < 5:
    print(count)
    count += 1
```

Output:

```
0
1
2
3
4
```

Use Cases:

- Repeating actions until a certain condition changes.
- Continuously prompting a user for input until they provide valid data.
- Implementing retry logic for network requests.

### 3. **Nested Loops**

Python allows nesting of `for` and `while` loops. This is useful when you need to perform multiple levels of iteration, such as working with multidimensional data structures (like lists of lists).

Example:


for i in range(3):
    for j in range(2):
        print(f"i={i}, j={j}")
```

Output:

```
i=0, j=0
i=0, j=1
i=1, j=0
i=1, j=1
i=2, j=0
i=2, j=1
```

Use Cases:

- Iterating over two-dimensional arrays (e.g., matrices).
- Working with complex data structures that involve multiple levels of information.

4.`break` and `continue` Statements**

- `break`: Exits the loop prematurely, regardless of the loop's original termination condition.
    Example:
    for num in range(10):
        if num == 5:
            break
        print(num)
    ```

    Output:

    ```
    0
    1
    2
    3
    4
    ```

- `continue`: Skips the current iteration and proceeds to the next iteration of the loop.

    Example:

    for num in range(5):
        if num == 2:
            continue
        print(num)
    ```

    **Output:**

    ```
    0
    1
    3
    4
    ```

 Summary

- `for` loop: Best for iterating over a fixed set of items or a known number of iterations.
- `while` loop: Best for scenarios where the number of iterations is not known beforehand and depends on a condition.
- Nested loops: Used for working with multi-dimensional data structures or performing complex iterations.
- `break` and `continue`: Control flow within loops, allowing for more complex behavior.
