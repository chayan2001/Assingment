```python
df=pd.read_csv('amazon.csv')
```

```python
#question 1
df.rating=pd.to_numeric(df['rating'],errors='coerce')
category_rating=df.groupby('category') ['rating'].mean()
```

```python
# Create a figure and axes
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the distribution of discounted prices
sns.histplot(df['discounted_price'], ax=ax, label='Discounted Price', kde=True, color='blue')

# Plot the distribution of actual prices
sns.histplot(df['actual_price'], ax=ax, label='Actual Price', kde=True, color='orange')

# Set plot labels and title
ax.set_xlabel('Price')
ax.set_ylabel('Frequency')
ax.set_title('Distribution of Discounted vs. Actual Prices')

# Add a legend
ax.legend()

# To see the output, run the code.
```

```python
#Question 5
df['rating_count'] = pd.to_numeric(df['rating_count'].str.replace(',', ''), errors='coerce')
```

```python
df.groupby('product_name') ['rating_count'].sum().sort_values(ascending=False).head(5)
```

```python
#AmazonBasics Flexible Premium HDMI Cable (Black, 4K@60Hz, 18Gbps), 3-Foot this is most popular product
```

```python
product_popularity = df.groupby('product_name')['rating_count'].sum()
```

```python
most_popular_products = product_popularity.sort_values(ascending=False)
```
