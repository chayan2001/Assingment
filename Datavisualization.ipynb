```python
##seaborn library
#q1
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generating synthetic dataset with 1000 samples
n_samples = 1000

# Variable 1: Random data from normal distribution (mean=50, std=10)
var1 = np.random.normal(50, 10, n_samples)

# Variable 2: Random data from normal distribution with some correlation to var1
var2 = var1 * 0.75 + np.random.normal(0, 5, n_samples)

# Create a DataFrame for easier plotting with Seaborn
data = pd.DataFrame({
    'Variable 1': var1,
    'Variable 2': var2
})

# Create a scatter plot using Seaborn
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Variable 1', y='Variable 2', data=data, alpha=0.5, color='blue')
plt.title("Scatter Plot of Two Synthetic Variables using Seaborn")
plt.grid(True)
plt.show()
```

```python
#q2
random_data = np.random.normal(loc=50, scale=15, size=1000)  # Mean = 50, std = 15

# Create a histogram to visualize the distribution of the numerical variable
plt.figure(figsize=(8, 6))
sns.histplot(random_data, bins=30, kde=True, color='green')  # KDE line for smooth distribution curve
plt.title("Distribution of a Synthetic Numerical Variable")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
#q3
# Generate a dataset with categories and corresponding values
categories = ['A', 'B', 'C', 'D']
values = [np.random.randint(50, 150) for _ in range(4)]

data = pd.DataFrame({
    'Category': categories,
    'Value': values
})

# Bar plot to compare categories
plt.figure(figsize=(8, 6))
sns.barplot(x='Category', y='Value', data=data, palette='viridis')
plt.title("Comparison of Categories Based on Values")
plt.show()
```

```python
#q4
n_samples = 100
categories = np.random.choice(['A', 'B', 'C', 'D'], size=n_samples)
numerical_values = np.random.normal(50, 10, size=n_samples)

data = pd.DataFrame({
    'Category': categories,
    'Numerical Value': numerical_values
})

# Box plot to visualize the distribution of a numerical variable across categories
plt.figure(figsize=(8, 6))
sns.boxplot(x='Category', y='Numerical Value', data=data, palette='Set2')
plt.title("Distribution of Numerical Values Across Categories")
plt.show()
```

```python
##q5.
np.random.seed(42)
n_samples = 1000
feature_1 = np.random.normal(50, 10, n_samples)
feature_2 = feature_1 * 0.8 + np.random.normal(0, 5, n_samples)  # Correlated with feature_1
feature_3 = np.random.normal(60, 15, n_samples)

data = pd.DataFrame({
    'Feature 1': feature_1,
    'Feature 2': feature_2,
    'Feature 3': feature_3
})

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title("Correlation Matrix of Features")
plt.show()
```

```python
#q1
import plotly.graph_objects as go
import pandas as pd
import numpy as np

# Dataset
np.random.seed(30)
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}
df = pd.DataFrame(data)

# 3D Scatter plot
fig = go.Figure(data=[go.Scatter3d(x=df['X'], y=df['Y'], z=df['Z'], mode='markers')])
fig.update_layout(title="3D Scatter Plot", scene=dict(xaxis_title='X', yaxis_title='Y', zaxis_title='Z'))
fig.show()
```

```python
#q3
import plotly.express as px

# Dataset
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# Pivot data for heatmap
heatmap_data = df.pivot_table(index='Day', columns='Month', values='Sales')

# Heatmap plot
fig = px.imshow(heatmap_data, title="Sales Variation Across Months and Days", color_continuous_scale="Viridis")
fig.show()
```

```python
#q4
import plotly.graph_objects as go
import numpy as np
import pandas as pd

# Dataset
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# 3D Surface plot
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])
fig.update_layout(title="3D Surface Plot", scene=dict(xaxis_title='X', yaxis_title='Y', zaxis_title='Z'))
fig.show()
```

```python
#Bokeh
#q1
#Bokeh
#q1
from bokeh.plotting import figure, show
import numpy as np
from bokeh.io import output_notebook

# Prepare the output for the notebook
output_notebook()

# Create x-values from 0 to 10
x = np.linspace(0, 10, 100)
# Calculate y-values as the sine of x
y = np.sin(x)

# Create a Bokeh plot
# Use 'width' and 'height' instead of 'plot_width' and 'plot_height'
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)', width=800, height=400)
p.line(x, y, legend_label="sin(x)", line_width=2, color='blue')

# Show the plot
show(p)

```

```python
#q2
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np
import pandas as pd

# Prepare the output for the notebook
output_notebook()

# Generate random data
np.random.seed(42)  # Set seed for reproducibility
num_points = 50
data = {
    'x': np.random.rand(num_points) * 10,
    'y': np.random.rand(num_points) * 10,
    'sizes': np.random.randint(5, 20, num_points),
    'colors': np.random.choice(['red', 'blue', 'green', 'orange'], num_points)
}
df = pd.DataFrame(data)

# Create a Bokeh plot
p = figure(title="Scatter Plot with Random Data", width=600, height=400)

# Add scatter glyphs
p.scatter(
    x='x',
    y='y',
    source=df,  # Use the DataFrame as data source
    size='sizes',  # Vary marker size based on 'sizes' column
    color='colors',  # Vary marker color based on 'colors' column
    alpha=0.8  # Set marker transparency
)

# Show the plot
show(p)
```

```python
#q3

from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Prepare the output for the notebook
output_notebook()

# Fruit data
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a Bokeh plot
p = figure(x_range=fruits, title="Fruit Counts", width=600, height=400)

# Add bar glyphs
p.vbar(x=fruits, top=counts, width=0.8, color="navy", alpha=0.7)

# Customize the plot
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.xaxis.major_label_orientation = 1.2  # Rotate x-axis labels

# Show the plot
show(p)
```

```python
#q4
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Prepare the output for the notebook
output_notebook()

# Generate data for the histogram
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Create a Bokeh plot
p = figure(title="Histogram of Random Data", width=600, height=400)

# Add quad glyphs for the histogram bars
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
       fill_color="navy", line_color="white", alpha=0.7)

# Customize the plot
p.xaxis.axis_label = "Data Values"
p.yaxis.axis_label = "Frequency"

# Show the plot
show(p)
```

```python
#q5
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import LinearColorMapper, ColorBar

# Prepare the output for the notebook
output_notebook()

# Generate data for the heatmap
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Create a Bokeh plot
p = figure(title="Heatmap", width=600, height=400,
           x_range=(x.min(), x.max()), y_range=(y.min(), y.max()),
           toolbar_location=None)

# Define a color mapper
color_mapper = LinearColorMapper(palette="Viridis256", low=data_heatmap.min(), high=data_heatmap.max())

# Add rect glyphs for the heatmap cells
p.rect(x=xx.flatten(), y=yy.flatten(), width=x[1] - x[0], height=y[1] - y[0],
       fill_color={'field': 'value', 'transform': color_mapper},
       source={'value': data_heatmap.flatten()})

# Add a color bar
color_bar = ColorBar(color_mapper=color_mapper, label_standoff=12,
                    border_line_color=None, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```
