```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('spotify.csv')
```

```python
df.head()
```

```python
#question 1
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
#question 2
plt.hist(df['Popularity'])
plt.title('Distribution of Track Popularity')
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.show()
```

```python
#question3
plt.scatter(x='Duration (ms)',y='Popularity',data=df)
```

```python
#question 4
df.head()
```

```python
df.groupby('Artist')['Track Name'].value_counts().idxmax()
```

```python
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
sns.countplot(x='Artist', data=df)
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.title('Number of Tracks per Artist')
plt.xlabel('Artist')
plt.ylabel('Number of Tracks')
plt.show()
```

```python
#question 5
df.groupby('Track Name')['Popularity'].max().sort_values(ascending=True).head()
```

```python
df.groupby('Artist')['Popularity'].mean().sort_values(ascending=False).head()
```

```python
#question 8
# Assuming 'df' is your pandas DataFrame

# Select the numerical columns you want to visualize
numerical_cols = ['Popularity', 'Duration (ms)']  # Add more columns as needed

# Create the pair plot
sns.pairplot(df[numerical_cols])
plt.show()
```

```python
sns.violinplot(x='Artist', y='Duration (ms)', data=df)
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.title('Track Duration Variation Across Artists (Violin Plot)')
plt.xlabel('Artist')
plt.ylabel('Duration (ms)')
plt.show()
```

```python
plt.figure(figsize=(12, 6))  # Adjust figure size as needed
sns.violinplot(x='Artist', y='Popularity', data=df)
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.title('Track Popularity Distribution Across Artists (Violin Plot)')
plt.xlabel('Artist')
plt.ylabel('Popularity')
plt.show()
```
