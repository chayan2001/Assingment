````python
'''
#iii. Explain the difference between instance methods and class methods. Provide an example of each.
#ans.In Python, both instance methods and class methods are used to define functions within a class, but they serve different purposes and have different behaviors.

### Instance Methods
- **Definition**: Instance methods are functions defined within a class that operate on an instance of that class. They take `self` as the first parameter, which represents the instance on which the method is called.
- **Purpose**: Used to access or modify the attributes of the instance. They can access both instance-level and class-level attributes.
- **Usage**: Typically used to perform operations that involve the specific instance of the class.

**Example of an Instance Method:**
```python
class Dog:
    def __init__(self, name):
        self.name = name

    def bark(self):
        print(f"{self.name} says Woof!")

# Creating an instance of the Dog class
my_dog = Dog("Buddy")
my_dog.bark()  # Output: Buddy says Woof!
```
Here, `bark` is an instance method because it uses `self` to refer to the instance `my_dog` and accesses its `name` attribute.

### Class Methods
- **Definition**: Class methods are functions that belong to the class itself rather than an instance of the class. They take `cls` as the first parameter, which represents the class.
- **Purpose**: Used to perform operations that are related to the class as a whole, rather than any specific instance. They can modify class-level attributes or provide utility functions that are related to the class.
- **Usage**: Typically used when the method needs to access or modify the class state, or when it doesn't need to interact with instance-specific data.

**Example of a Class Method:**
```python
class Dog:
    species = "Canine"  # Class attribute

    def __init__(self, name):
        self.name = name

    @classmethod
    def get_species(cls):
        return cls.species

# Accessing the class method without creating an instance
print(Dog.get_species())  # Output: Canine
```
In this example, `get_species` is a class method. It uses `cls` to access the class attribute `species` without needing an instance of the `Dog` class.

### Key Differences:
1. **Instance Methods**:
   - Operate on individual instances of a class.
   - Use `self` as the first parameter.
   - Can access and modify instance attributes and methods.

2. **Class Methods**:
   - Operate on the class itself, not instances.
   - Use `cls` as the first parameter.
   - Can access and modify class attributes and class-level methods.
   - Decorated with `@classmethod`.
'''
````

````python
'''
iv.How does Python implement method overloading? Give an example.
ans.Python does not support traditional method overloading as seen in languages like Java or C++, where you can define multiple methods with the same name but different parameters. In Python, defining multiple methods with the same name will override the previous ones.

However, you can achieve a similar effect by using default arguments, variable-length arguments (`*args` and `**kwargs`), or by manually checking the types and number of arguments within a single method.

### Implementing Method Overloading Using Default Arguments
You can use default values for parameters to handle different cases.

```python
class Calculator:
    def add(self, a, b=0, c=0):
        return a + b + c

calc = Calculator()
print(calc.add(5))        # Output: 5 (a = 5, b = 0, c = 0)
print(calc.add(5, 10))    # Output: 15 (a = 5, b = 10, c = 0)
print(calc.add(5, 10, 15)) # Output: 30 (a = 5, b = 10, c = 15)
```

In this example, the `add` method behaves differently depending on the number of arguments provided.

### Implementing Method Overloading Using `*args` and `**kwargs`
You can use `*args` to accept a variable number of positional arguments and `**kwargs` for keyword arguments.

```python
class Calculator:
    def add(self, *args):
        return sum(args)

calc = Calculator()
print(calc.add(5))            # Output: 5
print(calc.add(5, 10))        # Output: 15
print(calc.add(5, 10, 15))    # Output: 30
print(calc.add(1, 2, 3, 4))   # Output: 10
```

In this case, `add` can take any number of arguments, and it sums them all up.

### Implementing Method Overloading Using Type Checking
You can check the types and number of arguments within a single method and perform different operations based on that.

```python
class Calculator:
    def add(self, a, b):
        if isinstance(a, str) or isinstance(b, str):
            return str(a) + str(b)
        else:
            return a + b

calc = Calculator()
print(calc.add(5, 10))           # Output: 15 (Integer addition)
print(calc.add("Hello", "World")) # Output: HelloWorld (String concatenation)
print(calc.add(5, " Apples"))     # Output: 5 Apples (String concatenation)
```

Here, the `add` method changes its behavior based on the types of the arguments.

### Summary
Python does not natively support method overloading by defining multiple methods with the same name and different signatures. Instead, you can use techniques like default arguments, `*args`/`**kwargs`, or type checking within a single method to achieve similar functionality.
'''
````

````python
'''
v.What are the three types of access modifiers in Python? How are they denoted?
ans.In Python, there are three types of access modifiers used to define the visibility and accessibility of class attributes and methods. Although Python does not have strict access control like some other languages (e.g., `private`, `protected`, and `public` in Java), it uses a naming convention to indicate the intended access level. These access modifiers are:

### 1. **Public**
- **Denoted By**: Attributes and methods are public by default and can be accessed from inside or outside the class.
- **Notation**: No leading underscores.

**Example:**
```python
class MyClass:
    def __init__(self):
        self.public_var = "I am a public variable"

    def public_method(self):
        print("I am a public method")

obj = MyClass()
print(obj.public_var)  # Output: I am a public variable
obj.public_method()    # Output: I am a public method
```

### 2. **Protected**
- **Denoted By**: A single leading underscore `_` before the attribute or method name.
- **Purpose**: Indicates that the attribute or method is intended for internal use only and should not be accessed directly from outside the class. However, this is just a convention and not enforced by the Python language.
- **Notation**: `_protected_var`

**Example:**
```python
class MyClass:
    def __init__(self):
        self._protected_var = "I am a protected variable"

    def _protected_method(self):
        print("I am a protected method")

obj = MyClass()
print(obj._protected_var)  # Output: I am a protected variable
obj._protected_method()    # Output: I am a protected method
```

In this example, `_protected_var` and `_protected_method()` are intended to be used within the class or its subclasses, but they can still be accessed from outside if needed.

### 3. **Private**
- **Denoted By**: A double leading underscore `__` before the attribute or method name.
- **Purpose**: Used to restrict access to class attributes and methods from outside the class. Python performs name mangling, which changes the name of the attribute in a way that makes it harder to access directly.
- **Notation**: `__private_var`

**Example:**
```python
class MyClass:
    def __init__(self):
        self.__private_var = "I am a private variable"

    def __private_method(self):
        print("I am a private method")

obj = MyClass()
# The following lines would raise an AttributeError
# print(obj.__private_var)
# obj.__private_method()

# Accessing the private members using name mangling
print(obj._MyClass__private_var)  # Output: I am a private variable
obj._MyClass__private_method()    # Output: I am a private method
```

In this example, `__private_var` and `__private_method()` are private members. They cannot be accessed directly using their original names, but can still be accessed through name mangling (`_ClassName__attributeName`).

### Summary
- **Public**: No underscores. Accessible from anywhere.
- **Protected**: Single underscore (`_`). Intended for internal use, but accessible from outside.
- **Private**: Double underscore (`__`). Name-mangled to prevent direct access, intended for internal use only.
'''
````

````python
'''
vii.7. What is the Method Resolution Order (MRO) in Python? How can you retrieve it programmatically?
ans.The **Method Resolution Order (MRO)** in Python is the sequence in which Python looks for a method or attribute in a class hierarchy when there are multiple classes involved, such as in multiple inheritance. MRO is crucial in determining which method is called when a method is invoked on an object, especially when dealing with complex inheritance trees.

### Key Concepts of MRO:
1. **Linearization of Classes**: MRO linearizes the class hierarchy, ensuring a consistent and predictable order in which classes are searched.
2. **C3 Linearization Algorithm**: Python uses the C3 linearization algorithm to compute the MRO, which resolves the order in such a way that it respects the inheritance hierarchy and avoids conflicts like the "diamond problem".
3. **`super()` Function**: The MRO determines how the `super()` function delegates method calls in a hierarchy.

### Example of MRO:
Let's consider an example of multiple inheritance:

```python
class A:
    def method(self):
        print("Method from A")

class B(A):
    def method(self):
        print("Method from B")

class C(A):
    def method(self):
        print("Method from C")

class D(B, C):
    pass

d = D()
d.method()  # Output: Method from B
```

In this example, `D` inherits from both `B` and `C`, which in turn inherit from `A`. When `d.method()` is called, Python follows the MRO to determine that the `method` from class `B` should be called first.

### How to Retrieve MRO Programmatically

You can retrieve the MRO in two ways:

1. **Using the `__mro__` Attribute**: This attribute returns a tuple representing the MRO of the class.
2. **Using the `mro()` Method**: This method returns a list representing the MRO of the class.

#### 1. Using `__mro__` Attribute
The `__mro__` attribute of a class provides the MRO as a tuple of classes.

```python
print(D.__mro__)
# Output: (<class '__main__.D'>, <class '__main__.B'>, <class '__main__.C'>, <class '__main__.A'>, <class 'object'>)
```

#### 2. Using `mro()` Method
The `mro()` method of a class returns the MRO as a list of classes.

```python
print(D.mro())
# Output: [<class '__main__.D'>, <class '__main__.B'>, <class '__main__.C'>, <class '__main__.A'>, <class 'object'>]
```

In both cases, the MRO for class `D` is:
`D -> B -> C -> A -> object`.
This means Python will search for the method in this order, starting from `D` and ending at the base `object` class.

### Why MRO Matters
Understanding the MRO is essential in scenarios involving multiple inheritance because it:
- Prevents ambiguity in method calls.
- Ensures that each class's method is called in a predictable order.
- Helps in correctly utilizing the `super()` function, which relies on the MRO to navigate through the inheritance chain.

### Summary
- **MRO** defines the order in which methods and attributes are resolved in a class hierarchy.
- **Retrieve MRO**: Using the `__mro__` attribute or `mro()` method.
- **MRO Calculation**: Python uses the C3 linearization algorithm to compute the MRO.
'''
````

````python
'''
viii.Create an abstract base class `Shape` with an abstract method `area()`. Then create two subclasses
`Circle` and `Rectangle` that implement the `area()` method.
ans.To create an abstract base class in Python, you use the `abc` (Abstract Base Classes) module. An abstract base class is a class that cannot be instantiated directly and typically includes one or more abstract methods that must be implemented by subclasses.

Here's how to create an abstract base class `Shape` with an abstract method `area()`, and then create two subclasses `Circle` and `Rectangle` that implement the `area()` method:

### Implementation:

```python
from abc import ABC, abstractmethod
import math

# Abstract Base Class
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

# Subclass Circle implementing the area() method
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

# Subclass Rectangle implementing the area() method
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

# Creating objects of Circle and Rectangle
circle = Circle(5)  # Radius = 5
rectangle = Rectangle(4, 6)  # Width = 4, Height = 6

# Calculating areas
print(f"Area of Circle: {circle.area()}")  # Output: 78.53981633974483
print(f"Area of Rectangle: {rectangle.area()}")  # Output: 24
```

### Explanation:
1. **Abstract Base Class (`Shape`)**:
   - The `Shape` class inherits from `ABC` (a helper class that allows the use of the `abstractmethod` decorator).
   - The `area()` method is decorated with `@abstractmethod`, meaning it must be implemented by any non-abstract subclass of `Shape`.

2. **Subclass `Circle`**:
   - The `Circle` class inherits from `Shape` and implements the `area()` method.
   - The `area()` method calculates the area of the circle using the formula `π * radius^2`.

3. **Subclass `Rectangle`**:
   - The `Rectangle` class also inherits from `Shape` and implements the `area()` method.
   - The `area()` method calculates the area of the rectangle using the formula `width * height`.

4. **Creating Instances and Calculating Area**:
   - Instances of `Circle` and `Rectangle` are created with specific dimensions.
   - The `area()` method is called on these instances, printing the respective areas.

### Key Points:
- Abstract base classes (ABCs) in Python are used to define interfaces and ensure that derived classes implement specific methods.
- The `abc` module and the `@abstractmethod` decorator are used to create abstract methods that subclasses must implement.
- Abstract classes cannot be instantiated directly and must be subclassed to be used.
'''
````

````python
'''
xii.Create a decorator that measures and prints the execution time of a function.
ans.A decorator in Python is a function that wraps another function to modify or enhance its behavior without changing its code. We can create a decorator to measure and print the execution time of any function it wraps by using the `time` module to record the start and end times of the function's execution.

### Implementation of the Decorator

```python
import time

def execution_time_decorator(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()  # Record the start time
        result = func(*args, **kwargs)  # Call the function
        end_time = time.time()  # Record the end time
        execution_time = end_time - start_time  # Calculate the execution time
        print(f"Execution time of '{func.__name__}': {execution_time:.6f} seconds")
        return result  # Return the result of the original function
    return wrapper

# Example usage of the decorator

@execution_time_decorator
def example_function(n):
    """A sample function that runs a loop to simulate work."""
    total = 0
    for i in range(n):
        total += i
    return total

# Calling the decorated function
result = example_function(1000000)
```

### Explanation:

1. **`execution_time_decorator` Function**:
   - This function is the decorator itself. It takes a function `func` as an argument.
   - Inside it, a nested `wrapper` function is defined. This function will wrap around the original function `func`.

2. **`wrapper` Function**:
   - **Start Time**: The `wrapper` function records the current time using `time.time()` before calling the original function. This is stored in `start_time`.
   - **Function Call**: It then calls the original function `func` with its arguments (`*args` and `**kwargs`) and stores the result in `result`.
   - **End Time**: After the function has finished executing, the `wrapper` records the end time in `end_time`.
   - **Execution Time Calculation**: The execution time is calculated by subtracting `start_time` from `end_time`.
   - **Print Execution Time**: The execution time is printed in seconds with six decimal places.
   - **Return Result**: Finally, the `wrapper` returns the result of the original function.

3. **Decorating a Function**:
   - The `@execution_time_decorator` syntax is used to decorate `example_function`.
   - When `example_function(1000000)` is called, it is first wrapped by the `execution_time_decorator` which measures and prints its execution time.

### Example Output:

```plaintext
Execution time of 'example_function': 0.034512 seconds
```

This indicates that the `example_function` took approximately 0.034512 seconds to execute. The actual time will vary depending on the system and the value of `n`.
'''
````

````python
'''
xii.Explain the concept of the Diamond Problem in multiple inheritance. How does Python resolve it?
ans.The **Diamond Problem** is a well-known issue in object-oriented programming that occurs when a class inherits from two classes that both derive from a single base class. This creates an ambiguity in the method resolution order (MRO) because the subclass has two paths to the base class.

### Understanding the Diamond Problem

Let's visualize the problem with an example:

```
     A
   /   \
  B     C
   \   /
     D
```

- **Class `A`**: The top base class.
- **Classes `B` and `C`**: Both inherit from `A`.
- **Class `D`**: Inherits from both `B` and `C`.

In this structure:
- `D` has two potential paths to reach `A`: `D -> B -> A` and `D -> C -> A`.
- If `A` has a method called `some_method()`, and `D` calls `some_method()`, it's unclear whether Python should call `A`'s implementation through `B` or through `C`. This is the essence of the diamond problem: **ambiguity in method resolution**.

### Python's Resolution: C3 Linearization (Method Resolution Order - MRO)

Python resolves the Diamond Problem using a specific algorithm called **C3 Linearization**. This algorithm ensures a consistent and predictable order for method resolution, following these rules:

1. **Depth-First Search (DFS)**: Python first performs a DFS to create a linear order of classes.
2. **Left-to-Right Ordering**: When resolving methods, the leftmost base class is considered first, following the order of inheritance defined in the class declaration.
3. **Avoiding Repetition**: The method from a base class is called only once, even if there are multiple paths to it.

Python computes the MRO in such a way that it respects the hierarchy and prevents calling the same method multiple times due to multiple inheritance paths.

### Example: Diamond Problem in Python

```python
class A:
    def show(self):
        print("Method from class A")

class B(A):
    def show(self):
        print("Method from class B")

class C(A):
    def show(self):
        print("Method from class C")

class D(B, C):
    pass

d = D()
d.show()
```

### Output:
```
Method from class B
```

### Explanation:

1. **MRO Calculation**:
   - For class `D`, the MRO is calculated as `[D, B, C, A, object]`.
   - This means that Python will look for the `show()` method in `D` first, then in `B`, then in `C`, and finally in `A`.

2. **Resolution**:
   - When `d.show()` is called, Python follows the MRO and finds the `show()` method in class `B` first, so it calls `B`'s implementation.
   - If `B` did not have a `show()` method, Python would continue to `C`, then to `A`.

3. **Preventing Ambiguity**:
   - The MRO ensures that there's no ambiguity even in complex inheritance structures, as it provides a deterministic order in which classes are searched.

### How to View the MRO:

You can view the MRO of a class using the `mro()` method or the `__mro__` attribute:

```python
print(D.mro())  # Output: [<class '__main__.D'>, <class '__main__.B'>, <class '__main__.C'>, <class '__main__.A'>, <class 'object'>]
print(D.__mro__)  # Output: (<class '__main__.D'>, <class '__main__.B'>, <class '__main__.C'>, <class '__main__.A'>, <class 'object'>)
```

### Summary

- **Diamond Problem**: Occurs in multiple inheritance when there is ambiguity in which method to call due to a class having multiple paths to a base class.
- **Python's Solution**: Uses the C3 Linearization algorithm to compute a consistent MRO that respects the hierarchy and prevents ambiguity.
- **MRO**: Defines the order in which classes are searched for methods and attributes, ensuring that the diamond problem is resolved correctly.
'''
````

````python
'''
xiv.Write a class method that keeps track of the number of instances created from a class.
ans.To create a class method that keeps track of the number of instances created from a class, we can use a class attribute to store the count and a class method to access or display this count. The class attribute is shared across all instances of the class, making it the perfect place to store such information.

Here's how you can implement this:

```python
class InstanceCounter:
    # Class attribute to keep track of the number of instances
    instance_count = 0

    def __init__(self):
        # Increment the class attribute each time a new instance is created
        InstanceCounter.instance_count += 1

    # Class method to get the number of instances
    @classmethod
    def get_instance_count(cls):
        return f"Number of instances created: {cls.instance_count}"

# Creating instances of InstanceCounter
obj1 = InstanceCounter()
obj2 = InstanceCounter()
obj3 = InstanceCounter()

# Using the class method to get the instance count
print(InstanceCounter.get_instance_count())  # Output: Number of instances created: 3

# Creating another instance
obj4 = InstanceCounter()

# Checking the updated instance count
print(InstanceCounter.get_instance_count())  # Output: Number of instances created: 4
```

### Explanation:

1. **Class Attribute `instance_count`**:
   - The class attribute `instance_count` is defined to keep track of the number of instances created from the `InstanceCounter` class.
   - This attribute is shared among all instances of the class.

2. **`__init__` Method**:
   - The `__init__` method is the constructor method in Python, called whenever a new instance of the class is created.
   - Each time an instance is created, `InstanceCounter.instance_count` is incremented by 1.

3. **Class Method `get_instance_count`**:
   - The class method `get_instance_count` is defined using the `@classmethod` decorator.
   - It takes `cls` as its first argument (instead of `self`), which refers to the class itself, allowing access to class attributes and other class methods.
   - This method returns the current value of `instance_count` and can be called on the class itself (e.g., `InstanceCounter.get_instance_count()`).

4. **Creating Instances and Checking Count**:
   - As new instances (`obj1`, `obj2`, etc.) are created, the `instance_count` is updated.
   - The `get_instance_count` method returns the total number of instances created so far, demonstrating that the class method and attribute are working as expected.

This pattern is useful when you need to track or limit the number of instances created from a class, such as in scenarios where you want to enforce singleton-like behavior or monitor object creation in a system.
'''
````

````python
'''
xv.Implement a static method in a class that checks if a given year is a leap year
ans.To implement a static method in a class that checks if a given year is a leap year, you can follow these steps:

1. Define a class.
2. Create a static method that performs the leap year check.
3. Use the method to determine if a specified year is a leap year.

### Leap Year Check Rules

A year is a leap year if:
- It is divisible by 4.
- However, if the year is divisible by 100, it is not a leap year, unless:
- The year is also divisible by 400, in which case it is a leap year.

### Implementation

Here's how you can implement this in Python:

```python
class Year:
    @staticmethod
    def is_leap_year(year):
        """Check if the given year is a leap year."""
        if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
            return True
        return False

# Example usage
year_to_check = 2024
if Year.is_leap_year(year_to_check):
    print(f"{year_to_check} is a leap year.")
else:
    print(f"{year_to_check} is not a leap year.")

# Check another year
year_to_check = 1900
if Year.is_leap_year(year_to_check):
    print(f"{year_to_check} is a leap year.")
else:
    print(f"{year_to_check} is not a leap year.")
```

### Explanation:

1. **Class Definition (`Year`)**:
   - The class `Year` is defined to encapsulate the leap year checking functionality.

2. **Static Method (`is_leap_year`)**:
   - The `@staticmethod` decorator is used to define the `is_leap_year` method, which does not depend on any instance or class-specific data.
   - This method takes a single parameter, `year`, and checks if it meets the conditions for being a leap year based on the rules outlined above.

3. **Leap Year Logic**:
   - The method uses a simple conditional statement to determine if the year is a leap year and returns `True` or `False` accordingly.

4. **Example Usage**:
   - The static method can be called directly on the class without needing to create an instance (`Year.is_leap_year(year)`).
   - The example checks both 2024 (a leap year) and 1900 (not a leap year), demonstrating how the method works.

This implementation provides a straightforward way to check for leap years using a static method, making it easy to use without requiring an object instance.
'''
````
