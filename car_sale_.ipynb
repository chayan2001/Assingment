```python
#question 6
gender_income_price = df.groupby(['Gender', 'Annual Income'])['Price ($)'].mean().reset_index()
```

```python
heatmap_data = gender_income_price.pivot(index="Gender", columns="Annual Income", values="Price ($)")
sns.heatmap(heatmap_data, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title('Average Selling Price by Gender and Annual Income')
plt.show()
```
